```python
from docx import Document  # Documentos de Word
import openpyxl  # Documentos de Excel
import fitz  # Documentos PDF

def abrir_archivo_word(archivo):
    try:
        document = Document(archivo)
        print("Word:")
        for paragraph in document.paragraphs:
            print(paragraph.text)
    except Exception as e:
        print(f"No se pudo abrir el archivo Word: {e}")

def abrir_archivo_excel(archivo):
    try:
        libro = openpyxl.load_workbook(archivo)
        hoja = libro.active
        print("Excel:")
        for row in hoja.iter_rows(values_only=True):
            print(", ".join(map(str, row)))
    except Exception as e:
        print(f"No se pudo abrir el archivo Excel: {e}")

def abrir_archivo_pdf(archivo):
    try:
        pdf_document = fitz.open(archivo)
        print("PDF:")
        for page_num in range(len(pdf_document)):
            page = pdf_document[page_num]
            text = page.get_text()
            print(text)
    except Exception as e:
        print(f"No se pudo abrir el archivo PDF: {e}")

word = "LinksImagenes.docx"
excel = "CronogramaActividades.xlsx"
pdf = "infografia_patrones.pdf"

abrir_archivo_word(word)
abrir_archivo_excel(excel)
abrir_archivo_pdf(pdf)
```
